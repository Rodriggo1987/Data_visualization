```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import bar_chart_race as bcr #pip install bar_chart_race    ///    pip install ffmpeg // pip install ffmpeg-python
```

```python
# Dados de Vendas
data = {
    'Month': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
    'Product A Sales': [117522, 85275, 83564, 84110, 78656, 79352, 78654, 80014, 67560, 567782, 578041, 602280],
    'Product B Sales': [153342, 150887, 150222, 146647, 151786, 148740, 154047, 151366, 152004, 181796, 192452, 201947]
}

df = pd.DataFrame(data)
```

```python


# Definir as barras
barWidth = 0.35

# Definir as posições das barras no eixo x
r1 = range(len(df['Month']))   #representa a posição das barras no eixo x para o primeiro conjunto de barras (nesse caso, as vendas do Produto A).
r2 = [x + barWidth for x in r1] #representa a posição das barras no eixo x para o segundo conjunto de barras (nesse caso, as vendas do Produto B).

# Criar o gráfico de colunas
plt.bar(r1, df['Product A Sales'], color='blue', width=barWidth, edgecolor='grey', label='Produto A')
plt.bar(r2, df['Product B Sales'], color='orange', width=barWidth, edgecolor='grey', label='Produto B')

# Adicionar rótulos e título
plt.xlabel('Meses', fontweight='bold', fontsize=12)
plt.ylabel('Vendas', fontweight='bold', fontsize=12)
plt.xticks([r + barWidth for r in range(len(df['Month']))], df['Month'], rotation=45)
plt.title('Vendas - 2023', fontweight='bold')

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.show()

#barWidth =  largura da barra
```

```python
# Definir as barras
barWidth = 0.35

# Definir as posições das barras no eixo x
r1 = range(len(df['Month']))   #representa a posição das barras no eixo x para o primeiro conjunto de barras (nesse caso, as vendas do Produto A).
r2 = [x + barWidth for x in r1] #representa a posição das barras no eixo x para o segundo conjunto de barras (nesse caso, as vendas do Produto B).

# Criar o gráfico de colunas
plt.bar(r1, df['Product A Sales'], color='blue', width=barWidth, edgecolor='grey', label='Produto A')
plt.bar(r2, df['Product B Sales'], color='orange', width=barWidth, edgecolor='grey', label='Produto B')

# Adicionar rótulos e título
plt.xlabel('Meses de Maior Vendas', fontweight='bold')
plt.ylabel('Vendas', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(df['Month']))], df['Month'], rotation=45)
plt.title('Destaque de Vendas no Ano', fontweight='bold')
plt.xticks((9, 10, 11), [r'$\mathbf{\it{Outubro}}$', r'$\mathbf{\it{Novembro}}$', r'$\mathbf{\it{Dezembro}}$'], fontsize=12)


# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.show()

#barWidth =  largura da barra
```

```python
# Calcular o total de vendas de cada mês
df['Total Sales'] = df['Product A Sales'] + df['Product B Sales']

# Adicionar uma coluna com porcentagens
df['Percentage'] = (df['Product A Sales'] / df['Total Sales']) * 100

# Criar o gráfico de barras com porcentagens
plt.figure(figsize=(10, 6))
bars = plt.bar(df['Month'], df['Percentage'], color='skyblue', label='% De Vendas dos Produtos A e B')

# Adicionar porcentagens acima das barras
for bar, percentage in zip(bars, df['Percentage']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{percentage:.2f}%',
             ha='center', va='bottom')

# Adicionar rótulos e título
plt.xlabel('Meses', fontweight='bold')
plt.ylabel('Porcentagem de Produtos Vendidos (%)', fontweight='bold')
plt.title('Porcentagem de Produtos Vendidos por Mês', fontsize=15)
plt.legend()

# Exibir o gráfico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calcular o total de vendas de cada mês
df['Total Sales'] = df['Product A Sales'] + df['Product B Sales']

# Adicionar uma coluna com porcentagens
df['Percentage'] = (df['Product A Sales'] / df['Total Sales']) * 100

# Identificar o mês com as menores vendas
min_month = df.loc[df['Total Sales'].idxmin(), 'Month']

# Criar o gráfico de barras com porcentagens
plt.figure(figsize=(10, 6))
bars = plt.bar(df['Month'], df['Percentage'], color='skyblue')

# Destacar o mês com as menores vendas
for bar, month in zip(bars, df['Month']):
    if month == min_month:
        bar.set_color('red')  # Destacar a barra em vermelho

# Adicionar porcentagens acima das barras
for bar, percentage in zip(bars, df['Percentage']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{percentage:.2f}%',
             ha='center', va='bottom')

# Adicionar rótulos e título
plt.xlabel('Meses')
plt.ylabel('Porcentagem de Produtos Vendidos (%)')
plt.title('Percentagem de Menor venda no Ano')

# Exibir o gráfico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
